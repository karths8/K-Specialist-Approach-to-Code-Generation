```python
import json
import re
```

```python
json_file_path = "processed_seed_data_w_gen_asserts.json"
```

```python
json_file = open(json_file_path, 'r')
```

```python
data = json.load(json_file)
```

```python
def execute_python_code(code_string):
    try:
        exec(code_string, globals())
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        print(code_string)
        print()
        return e
```

```python
error_list=[]
correct=0
for datum in data:
    all_asserts = datum['asserts']
    gen_asserts = re.findall(r'assert .+ == .+', datum['gen_asserts'])
    all_asserts.extend(gen_asserts)

    if 'difficulty' not in datum:
        datum['difficulty'] = None
    
    err_cnt = 0
    errors = []
    correct = []
    for asrt in all_asserts:
        full_code = datum['code'] + '\n' + asrt
        error = execute_python_code(full_code)
        if error:
            err_cnt += 1
            if any(char.isalpha() for char in str(error)):
                errors.append(error)
            else:
                errors.append('Assert Failed: ' + asrt)
        else:
            correct.append(asrt)
    if err_cnt > 0:
        error_list.append(datum)
        datum['Error'] = errors
    datum['Total Asserts'] = len(all_asserts)
    datum['Failed Asserts'] = len(errors)
    datum['Passed Asserts'] = correct
```

```python
selected_data = [{"author": entry["author"], 
                  "UUID": entry["UUID"],
                  "question": entry["question"],
                  "examples": entry["examples"],
                  "link": entry["link"],
                  "difficulty": entry["difficulty"],
                  "categories": entry["categories"],
                  "code": entry["code"],
                  "asserts": entry["Passed Asserts"]
                 } for entry in data]
```

```python
with open("final_seed_data.json", 'w') as file:
    json.dump(selected_data, file)
```
