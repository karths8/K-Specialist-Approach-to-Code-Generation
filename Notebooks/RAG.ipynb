```python
import json
with open('../Data_files/seed_data_final_Nov18.json', 'r') as json_file:
    seed = json.load(json_file)
```

```python
categories = []
for i in seed:
    categories.extend(i['categories'])
```

```python
category_dict = {}
for i in seed:
    cat = i['categories'][0]
    if cat not in category_dict:
        category_dict[cat]=[]
    category_dict[cat].append(i)
```

```python
category_dict
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Count the occurrences of each string
counter = Counter(categories)

# Extract unique strings and their frequencies
labels, frequencies = zip(*counter.items())
plt.figure(figsize=(18, 12))
# Plotting a bar chart
plt.bar(labels, frequencies)

# Adding labels and title
plt.xlabel('Strings')
plt.ylabel('Frequency')
plt.title('String Frequency Bar Chart')

# Show the plot
plt.show()
```

```python
from itertools import permutations

all_permutations = []

for l in category_dict:
    ex_list = category_dict[l]
    fs_num = min(4,len(ex_list))
    perms = list(permutations(ex_list,fs_num))
    all_permutations.extend(perms)
    
print(len(all_permutations))
```

```python
def make_prompt_str(data):
    prompt_str=f'''Question:
        
{data['Question']}

Example / Explanation:

{data['Question Example/Explanation']}'''
    code_str = f'''\nCode:

{data['Method 1']} [/STOP]'''
    
    return prompt_str, code_str
```

```python
def make_chat_template(chat_type,data):
    if chat_type=='system':
        chat = [{"role": "system","content":data['input']}]
    elif chat_type=='sample':
        chat = [{"role": "user","content":data['input']},{"role": "assistant","content":data['output']}]
    return chat
```
