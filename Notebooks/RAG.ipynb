```python
import json
with open('../Data_files/final_seed_data.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
def get_cats(data):
    cats = {}
    for i in data:
        cat = i['categories'][0]
        if cat not in cats:
            cats[cat]=0
        cats[cat]+=1
    return cats
```

```python
from itertools import permutations

all_permutations = []

for l in category_dict:
    ex_list = category_dict[l]
    fs_num = min(3,len(ex_list))
    perms = list(permutations(ex_list,fs_num))
    all_permutations.extend(perms)
    
print(len(all_permutations))
```

```python
def make_few_shot(data):
    asserts = '\n'.join(data['asserts'])
    prompt_str=f'''Question:
        
{data['question']}'''
    output_str=f'''
Examples:

{data['examples']}

Code:

{data['code']}

Test cases:

{asserts}

[/STOP]'''
    
    return prompt_str, output_str
```

```python
def make_chat_template(chat_type,data):
    if chat_type=='system':
        chat = [{"role": "system","content":data['input']}]
    elif chat_type=='sample':
        chat = [{"role": "user","content":data['input']},{"role": "assistant","content":data['output']}]
    return chat
```

```python
from transformers import AutoTokenizer
def prompt_composer(data_with_perms):
    sys_prompt = {'input':'''
You are an assistant tasked to generate new data points from a few given ones. The data points will have a "question", some "Examples" that demonstrate input/output behaviour and a "Code" solution that confoms to the "question" and "Examples". You will also generate test cases or "assert" statements that are to be used to test the validity of the function in answering the question.
'''}
    prompt_list = []
    tokenizer = AutoTokenizer.from_pretrained('/workspace/CS762_Project/CodeLlama-34b-Python-hf')
    for i in data_with_perms:
        full_prompt = make_chat_template('system',sys_prompt)
        for ex in i:
            ip, op = make_few_shot(ex)
            d = {'input':ip, 'output':op}
            full_prompt.extend(make_chat_template('sample',d))
        full_prompt.append({"role": "user","content":"Question:"})
        prompt_list.append({'prompt':tokenizer.apply_chat_template(full_prompt, tokenize=False)[:-8],'examples':i})
    return prompt_list

prompt_list = prompt_composer(all_permutations)
```
