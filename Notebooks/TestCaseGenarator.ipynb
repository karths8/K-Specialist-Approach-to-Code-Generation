```python
testcase_template = """
Question:
{question}

Code:
{code}

Testcases:

TC1:
Input: {input}
Output: {output}
Assert Statement: {ass}

TC2:
"""
```

```python
testcase = testcase_template.format(question = "", code = "", input = "", output = "", ass = "")
```

```python
for index, row in df.iterrows():
    if row['Gen Asserts'] is None:
        testcase_gen_prompt = testcase_template.format(question = row['Question'], 
                                                       code = row['Method 1'], 
                                                       input = row['Input'], 
                                                       output = row['Output'], 
                                                       ass = row['Assert Statement 1'])
        response = openai.ChatCompletion.create(
            model='gpt-4',
            messages=[
                {"role": "system", "content": "You are an expert programmer that generates more test cases for a coding question. You are given the question and its corresponding code along with one test case. You must refer to the given data to generate more test cases. Make sure that there is no overlap between the Examples and test cases."},
                {"role": "user", "content": testcase_gen_prompt},
            ],
            temperature=0.7,
        )
        df.at[index, 'Gen Asserts'] = response.choices[0]["message"]["content"]
```
