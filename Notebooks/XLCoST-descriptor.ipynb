````python
import numpy as np

def calculate_quartiles(data):
    """
    Calculate quartiles and quartile breaks for a list of numbers.
    """
    # Sort the data
    sorted_data = np.sort(data)

    # Calculate quartiles using numpy's percentile function
    q1 = np.percentile(sorted_data, 25)
    q2 = np.percentile(sorted_data, 50)  # Median (Q2)
    q3 = np.percentile(sorted_data, 75)


    return q1, q2, q3

# Example usage
data = [12, 7, 3, 1, 6, 8, 11, 13, 15, 14, 19, 20]

q1, q2, q3 = calculate_quartiles([len(i['text']) for i in dataset_list])

# Print the results
print("Data:", data)
print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
````

````python
def make_example(example, n, test=False):
    if test:
        return {'user':f'[Title {n}]\n\n{example["text"]}\n\n[\Title {n}]'}
    else:
        assert_str = '\n'.join(example['asserts'])
        return {'user':f'[Title {n}]\n\n{example["title"]}\n\n[\Title {n}]', 'assistant':f'[Question {n}]\n\n{example["question"]}\n\n[\Question {n}]\n\n[Code {n}]\n\n{example["code"]}\n\n[\Code {n}]\n\n[Asserts {n}]\n\n{assert_str}\n\n[\Asserts {n}]'}
    
````

````python
def build_example_template(data_examples):
    examples = []
    for i in data_examples:
        examples.append({"role":"user", "content":i["user"]})
        examples.append({"role":"assistant", "content":i["assistant"]})
    return examples
````

````python
from copy import deepcopy
from sentence_transformers import SentenceTransformer, util

def get_few_shots(examples,data, n=2):
    print('Model loading!')
    model = SentenceTransformer('/workspace/CS762_Project/e5-large-v2')
    print('Embeddings loading!')
    examples = [e['title'] for e in examples]
    data = [d['text'] for d in data]
    example_embeddings = model.encode(examples)
    data_embeddings = model.encode(data)
    cosine_scores = util.cos_sim(data_embeddings, example_embeddings)
    print('Sorting!')
    sorted_cs_dict = {}
    for idx1 in range(len(cosine_scores)):
        cs_list = deepcopy(cosine_scores[idx1])
        cs_dict = {index: value.item() for index, value in enumerate(cs_list)}
        sorted_cs_dict[idx1] = {k: v for k, v in sorted(cs_dict.items(), key=lambda item: item[1], reverse=True) if v<0.999}
        sorted_cs_dict[idx1] = dict(list(sorted_cs_dict[idx1].items())[:n])
    return sorted_cs_dict
````

````python
def remove_extra_python(sample):
    python_pattern = '```python(.*)```'
    filtered = re.search(python_pattern, sample, re.DOTALL)
    filtered = filtered.group(1).strip() if filtered else sample
    return filtered
````

````python
import re
def extract_qca(sample):
    question_pattern = r'\[Question 3\](.*)\[\\Question 3\]'
    code_pattern = r'\[Code 3\](.*)\[\\Code 3\]'
    assert_pattern = r'\[Asserts 3\](.*)\[\\Asserts 3\]'
    question = re.search(question_pattern, sample, re.DOTALL)
    code = re.search(code_pattern, sample, re.DOTALL)
    asserts = re.search(assert_pattern, sample, re.DOTALL)
    question = question.group(1).strip() if question else None
    code = remove_extra_python(code.group(1).strip()) if code else None
    asserts = [i for i in remove_extra_python(asserts.group(1).strip()).split('\n') if i!=''] if asserts else None
    return question, code, asserts
````
