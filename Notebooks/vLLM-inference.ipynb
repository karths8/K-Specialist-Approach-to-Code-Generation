```python
import json
with open('/workspace/CS762_Project/Notebooks/xlcost_filtered_results_Dec12.json', 'r') as json_file:
    # Read the JSON data
    data = json.load(json_file)
```

```python
from vllm import LLM, SamplingParams
prompts = [i['question'] for i in data][:10]
sampling_params = SamplingParams(temperature=0.8, top_p=0.95,max_tokens=1000, stop=['def'])
llm = LLM(model="/workspace/CS762_Project/CodeLlama-7b-Python-hf", trust_remote_code=True)
outputs = llm.generate(prompts, sampling_params)

for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")

```

```python
for idx, request_output in enumerate(outputs):
    request_id = request_output.request_id
    prompt = request_output.prompt
    prompt_token_ids = request_output.prompt_token_ids
    output = request_output.outputs[0].text
    print(f'Prompt {idx}:\n\n{prompt}\n\nOutput {idx}:\n\n{output}')
```
